```python
def training_graph(X_train,degree,Y_train):
    Q_ML= bestQ(X_train,degree,Y_train)
    feature_matrix= Feature_Matrix_genrater(X_train,degree)
    plt.scatter(X_train,Q_ML.dot(feature_matrix.T))
    plt.scatter(X_train,Y_train, color="red")

def bestQ(X_train,degree,Y_train):
    feature_matrix= Feature_Matrix_genrater(X_train,degree)
    Q_ML= np.linalg.inv((feature_matrix).T.dot(feature_matrix)).dot(feature_matrix.T).dot(Y_train)
    Weight.append([Q_ML])
    print(Q_ML)
    return Q_ML

def Feature_Matrix_genrater(X_train,degree):
    feature_matrix= [ ]
    for i in range(len(X_train)) :
        feature_matrix.append(row_maker(X_train[i],degree))
    feature_matrix= np.array(feature_matrix)
    return feature_matrix

def row_maker(data,degree):
    poly=[]
    for i in range(degree+1):
        poly.append(pow(data,i))
    return poly

def error_funtion(X_train,Y_train,X_test,Y_test,degree):
    feature_matrix_test= Feature_Matrix_genrater(X_test,degree)
    Q_ML= bestQ(X_train,degree,Y_train)
    error= np.power((Y_test-Yfinal(X_train,degree,Q_ML)),2)
    l2_norm = np.sum(np.power((error),2))/error.size
    lists=[]
    lists.append(np.sqrt(l2_norm/error.size))
    lists.append(degree)
    return lists

def Yfinal(X_train,degree,Q_ML):
    feature_matrix_test= Feature_Matrix_genrater(X_test,degree)
    yfinal_f= Q_ML.dot(feature_matrix_test.T)
    yfinal_f=np.array(yfinal_f)
    return yfinal_f
Weight = []
```
